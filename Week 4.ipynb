## EXPLORATORY DATA ANALYSIS

```python
# Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

#### 1. Create a line plot of ZN and INDUS in the housing data.

	a. For ZN, use a solid green line. For INDUS, use a blue dashed line.
	b. Change the figure size to a width of 12 and height of 8.
	c. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.

```python
df = pd.read_csv("~/Downloads/boston_housing_data.csv")
df
```

```python
# Extract ZN and INDUS data
zn_data = df['ZN']
indus_data = df['INDUS']

# Create a line plot
plt.figure(figsize=(12, 8))
plt.plot(zn_data, color='green', label='ZN', linestyle='-', linewidth=2)
plt.plot(indus_data, color='blue', label='INDUS', linestyle='--', linewidth=2)
plt.style.use('bmh')

# Add labels and a legend
plt.xlabel('Data Points')
plt.ylabel('Data Values')
plt.legend()

# Show the plot
plt.show()
```

#### 2. Create a bar chart using col1 and col2 of dummy data.

	a. Give the plot a large title of your choosing.
	b. Move the legend to the lower-left corner.
	c. Do the same thing but with horizontal bars.
	d. Move the legend to the upper-right corner.

```python
# Create a bar chart with vertical bars
plt.figure(figsize=(8, 6))
plt.bar(df.index, df['CRIM'], label='CRIM', color='blue', alpha=0.6)
plt.bar(df.index, df['ZN'], label='ZN', color='green', alpha=0.6, bottom=df['CRIM'])

# Set a large title
plt.title('VERTICAL HOUSING BAR CHART')

# Move the legend to the lower-left corner
plt.legend(loc='lower left')

# Show the plot
plt.show()

# Create a bar chart with horizontal bars
plt.figure(figsize=(8, 6))
plt.barh(df.index, df['CRIM'], label='CRIM', color='magenta', alpha=0.6)
plt.barh(df.index, df['ZN'], label='ZN', color='orange', alpha=0.6, left=df['CRIM'])

# Set a large title
plt.title('HORIZONTAL HOUSING BAR CHART')

# Move the legend to the upper-right corner
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

#### 3. Create a histogram with pandas for using MEDV in the housing data.
	a. Set the bins to 20

```python
# Create a histogram for the 'MEDV' column with 20 bins
plt.figure(figsize=(8, 6))
plt.hist(df['MEDV'], bins=20, color='purple', alpha=0.7)

# Add labels and a title
plt.xlabel('MEDV Values')
plt.ylabel('Frequency')
plt.title('Histogram of MEDV')

# Show the histogram
plt.show()
```

#### 4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set the figure size
plt.figure(figsize=(10, 8))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")

# Set the title
plt.title("Correlation Heatmap for Boston Housing Data")

# Show the plot
plt.show()
```

```python
# Identify the columns with a positive correlation (replace 'col1' and 'col2' with actual column names)
x = df['TAX']
y = df['CRIM']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.6)

# Set axis labels
plt.xlabel('TAX')
plt.ylabel('CRIM')

# Set a title
plt.title('Scatter Plot of Two Heatmap Entries with High Positive Correlation')

# Show the plot
plt.show()
```
