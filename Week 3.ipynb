### NUMPY PRACTICE QUESTIONS

#### 1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
import numpy as np

# Define custom numpy arrays A and B
A = np.array([[1, 2, 3], [4, 5, 6]])

B = np.array([[7, 4, 9], [10, 11, 2]])

# Stack A and B vertically
stacked_vertically = np.vstack((A, B))

# Stack A and B horizontally
stacked_horizontally = np.hstack((A, B))

print("Stacked vertically:\n", stacked_vertically)
print("Stacked horizontally:\n", stacked_horizontally)
```

#### 2. Find common elements between A and B. [Hint : Intersection of two sets]

```python
common_elements = np.intersect1d(A, B)

print("Common elements between A and B:", common_elements)
```

#### 3. Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]

```python
# Use np.where() to find elements within the range
elements_in_range = A[np.where((A >= 5) & (A <= 10))]

print("Elements within the range [5, 10] in array A:")
print(elements_in_range)
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

iris_2d
```

```python
# Filtering for 3rd column > 1.5 and 1st column < 5.0
condition = (iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)

# Use the boolean mask to filter the rows
filtered_rows = iris_2d[condition]

print("Filtered rows of iris_2d:")
print(filtered_rows)
```

#### 3. How to get the rows of a dataframe with row sum > 100?

```python
df_2 = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))

df_2
```

```python
sum_rows = df_2.sum(axis=1)
sum_rows
```

```python
sum_greater_100 = df_2[sum_rows > 100]
print(sum_greater_100)
```
