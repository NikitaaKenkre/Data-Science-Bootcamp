#### Play with different parameter of decision trees and random forests and see the impact on train and test error

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X_test, Y_test, test_size=0.3, random_state=42)

# Varying max_depth for Decision Trees
for max_depth in [5, 10, 15, 20]:
    # Train Decision Tree Regressor
    decisiontree = DecisionTreeRegressor(max_depth=max_depth, min_samples_split=5)
    decisiontree.fit(X_train, Y_train)

    # Evaluate train and test errors
    mae_train = mean_absolute_error(Y_train, decisiontree.predict(X_train))
    mse_train = mean_squared_error(Y_train, decisiontree.predict(X_train))
    mae_test = mean_absolute_error(Y_test, decisiontree.predict(X_test))
    mse_test = mean_squared_error(Y_test, decisiontree.predict(X_test))

    print(f"Decision Tree (max_depth={max_depth}):")
    print("Train MAE:", mae_train, "Train MSE:", mse_train)
    print("Test MAE:", mae_test, "Test MSE:", mse_test)
    print()

# Varying n_estimators for Random Forests
for n_estimators in [50, 100, 150, 200]:
    # Train RandomForestRegressor model
    randomf = RandomForestRegressor(n_estimators=n_estimators, random_state=42)
    randomf.fit(X_train, Y_train)

    # Evaluate train and test errors
    mae_train = mean_absolute_error(Y_train, randomf.predict(X_train))
    mse_train = mean_squared_error(Y_train, randomf.predict(X_train))
    mae_test = mean_absolute_error(Y_test, randomf.predict(X_test))
    mse_test = mean_squared_error(Y_test, randomf.predict(X_test))

    print(f"Random Forest (n_estimators={n_estimators}):")
    print("Train MAE:", mae_train, "Train MSE:", mse_train)
    print("Test MAE:", mae_test, "Test MSE:", mse_test)
    print()
```
